```python
import matplotlib.pyplot as plt

np.set_printoptions(2)
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x, y, label="SIN WAVE", color="red", linestyle="dotted")
plt.xlabel("TIME")
plt.ylabel("AMPLITUDE")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. DATA (Your Logic) ---
x_months = np.arange(1, 13)
sales = np.cumsum(np.random.randint(5, 20, 12))
marketing_spend = np.random.uniform(10, 50, 50)
new_customers = 2 * marketing_spend + np.random.normal(0, 5, 50)
productivity = np.random.normal(75, 10, 500)
regions = ['North', 'South', 'East', 'West']
regional_sales = [250, 180, 310, 210]

# --- 2. THE GRID ---
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Executive Performance Dashboard", fontsize=20, fontweight='bold')

# [0, 0] Line Plot
axes[0, 0].plot(x_months, sales, marker='o', color='tab:blue', linewidth=2)
axes[0, 0].set_title("Cumulative Sales Growth")
axes[0, 0].set_ylabel("Total Sales ($)")

# [0, 1] Scatter Plot
axes[0, 1].scatter(marketing_spend, new_customers, color='tab:green', alpha=0.7)
axes[0, 1].set_title("Marketing Spend vs New Customers")
axes[0, 1].set_xlabel("Spend ($)")
axes[0, 1].set_ylabel("Customers")
axes[0, 1].grid(True, linestyle='--', alpha=0.5)

# [1, 0] Histogram
axes[1, 0].hist(productivity, bins=25, color='tab:orange', edgecolor="black", alpha=0.8)
axes[1, 0].set_title("Employee Productivity Distribution")
axes[1, 0].set_xlabel("Score")

# [1, 1] Bar Chart
axes[1, 1].bar(regions, regional_sales, color=['#3366cc', '#dc3912', '#ff9900', '#109618'])
axes[1, 1].set_title("Regional Sales Comparison")
axes[1, 1].set_ylabel("Units Sold")

# --- 4. CLEAN UP ---
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
